# Importing Required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

# Importing Data

```python
d=pd.read_csv('Iris.csv')
d.head()
```

# Dropping Useless Variables

```python
d.drop('Id',inplace=True,axis=1)
```

```python
d.info()
```

```python
sns.pairplot(d,hue='Species',kind='scatter')
```

# Preparing Data

```python
y=d['Species']
X=d.drop('Species',axis=1)
```

# Training the Model

```python
dtree=DecisionTreeClassifier()
dtree.fit(X,y)
```

# Visualizing Decision Tree

```python
features =list(d.columns[0:-1])
features
```

```python
#Visualization of decision tree classifier
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
n =plot_tree(dtree,feature_names = features,
           class_names = y,
           filled = True, fontsize= 14)
```
